```python
#Se añade una columna para calcular el número de muertos por minuto
x = df['Body_Count']/df['Length_Minutes']
df['Body_Count_per_length'] = df['Body_Count']/df['Length_Minutes']
#x.index(x.max())
#df.sort_values(by='x', ascending=False)
df.loc[df['Body_Count_per_length']==x.max()]
```

**Ejercicio 4**: ¿Nos estamos volviendo más violentos con el tiempo en lo que a cine se refiere? ¿Cuánta evidencia tienes de tu respuesta? Hint: Para poder responder bien a este ejercicio, serían interesantes dos gráficos: violencia por año y películas por año...

```python
#Se agrupa la cuenta de FIlms por año
df_grouped = df.groupby('Year').agg({'Film':'count'})
#df_grouped = df_grouped.agg({'Year': [("range(1900,2000)","mean")], 'POPESTIMATE2011': [("2011_sum","sum")]})
df_grouped.head(10)
#Pelîculas por años
df_grouped.plot()
```

**Ejercicio 5**: La columna de género está en un formato difícil de tratar: múltiples géneros para una película están concatenados con "|". Convierte esa columna a listas de género. Eg., "Horror|Sci-Fi" -> "["Horror", "Sci-Fi"]".

```python
#Para cada elemento de la columna Genre, se puede iterar y aplicar la función split para convertir la string en lista
df['Genre'] = df['Genre'].str.replace('|', ' ')
for i, value in df['Genre'].items():
    df['Genre'][i] = value.split()
```
