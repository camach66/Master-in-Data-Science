## ¿Cual es el código postal con precio medio más caro de madrid?
Tip: excluye zipcodes atípicos

```python
zipcode = df

zipcode['price'] = zipcode['price'].str.replace("$", "")
zipcode['price'] = zipcode['price'].str.replace(',', '')
zipcode['price'] = zipcode['price'].str.replace(' ', '')
zipcode = df.astype({"price": float})
zipcode.fillna('zipcode')
zipcode = zipcode.astype({"zipcode": str})

#Después de asegurar la columna price y zipcode como string, se escoge todas las strings que contengan 28
zipcode = zipcode[zipcode.zipcode.str.contains("28")]

#Se realiza la media y se saca el zipcode del valor más alto
final = zipcode.groupby('zipcode').agg({'price':'mean'})

final.sort_values(by='price', ascending=False).head(1)
```

28850 - Torrejón de Ardoz tiene el apartamento con el precio medio más caro de Madrid.
