El **objetivo final del ejercicio es obtener una lista ordenada de numeros primos felices**.

El código necesario para leer ambas listas ya lo tenemos, es el siguiente:

````python
def file_to_list(file_name):
    """
    Gets a filename and returns a list 
    where each element is one line of the file
    """
    with open(file_name) as f:
        lines = f.read().splitlines()
        return lines

happy_numbers = file_to_list("numeros_felices.txt")
print("Numeros felices: {}".format(happy_numbers))

prime_numbers = file_to_list("numeros_primos.txt")
print("Numeros primos: {}".format(prime_numbers))
````

````python
import unicodedata
#Función que saca la lista de entrada en la misma lista convertida con todos los elementos enteros y sin caracteres extra
def clean_list(file_list_elements):
    """
    Given a list of strings, converts such list a list of integers if possible
    """
    clean_file =[]
    #se guarda en clean_file cada elemento de la lista entrada, borrando cualquier caracter extra (strip)
    #y añadiendo después de la conversión a entero cada elemento
    for integer in file_list_elements:
        try:
                integer = integer.strip()
                intelement = int(integer)
                clean_file.append(intelement)
        except:
                pass    
    return clean_file
````

````python
def file_to_clean_list(file_name):
    """
    Given a filename with a number per line, returns such numbers in a list 
    """
    with open(file_name) as f:
#Se abre el archivo y se lee, con splitlines cada elemento entero se posiciona ordenadamente en la lista lines
        lines = f.read().splitlines()
#A lines, que ya tiene una lista con todos los números, se añade el filtro para remover caracteres y convertir todos 
#los elementos a enteros
        lines=clean_list(lines)
        return lines
````

3) Elimina de esa estructura aquellas palabras que son comunes (conocidas como _stop words_). El siguiente bloque de código da un set de esas palabras en `spa_stop_words`.


````python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
spa_stop_words=set(stopwords.words('spanish'))
print(spa_stop_words)
````

````python
def remove_stop_words(words_and_repetitions, stop_words):
    """
    Removes the `stop_words` from `words_and_repetitions`
    """
#Primero se controla las keys en el bucle de las palabras prohibidas, y se compara con el dict a evaluar
    for word in stop_words:
        #comparar con el dict a evaluar
        if word in words_and_repetitions:
            #Si se encuentra la palabra prohibida, se borra del dict con la función del
            del words_and_repetitions[word]
        else:
            pass

    return words_and_repetitions



#prueba
#remove_stop_words(z, spa_stop_words)
````
